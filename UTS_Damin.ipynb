```python
# Assuming df is your DataFrame
# Assuming you want to perform one-hot encoding on categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns

# Perform one-hot encoding using pandas get_dummies
df_encoded = pd.get_dummies(df, columns=categorical_columns, dummy_na=False)

# Print the resulting DataFrame with one-hot encoded features
print(df_encoded.head())
```

```python
import seaborn as sns

correlation_matrix = df_encoded.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

**Analisis**
Gambar menunjukkan matriks korelasi, yang menunjukkan hubungan linier antara variabel. Nilai-nilai dalam matriks berkisar antara -1 hingga 1, dengan nilai-nilai lebih dekat ke 1 menunjukkan hubungan positif yang kuat, nilai-nilai lebih dekat ke -1 menunjukkan hubungan negatif yang kuat, dan nilai-nilai lebih dekat ke 0 menunjukkan tidak adanya hubungan.

1.   Jumlah Penduduk: Dalam gambar diatas jumlah penduduk berkorelasi positif kuat dengan jumlah_rumah dan jenis_rumahtembok
2.   Luas Wilayah: Sedangkan dalam luas wilayah berkorelasi positif kuat dengan jml_penduduk dan usia
3.   Jumlah RS: Dalam korelasi matrik datanya menjadi kosong dikarenakan data hanya berisi 1



```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Assuming df_encoded is your DataFrame with encoded features
X = df_encoded.select_dtypes(include=['number'])  # Select only numerical features

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA(n_components=2)  # Reduce to 2 principal components
X_pca = pca.fit_transform(X_scaled)

# Perform K-Means clustering on the PCA-reduced data
kmeans = KMeans(n_clusters=3)  # Choose the number of clusters
kmeans.fit(X_pca)

# Add cluster labels to the DataFrame
df_encoded['cluster'] = kmeans.labels_

# Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans.labels_, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-Means Clustering with PCA')
plt.show()
```
